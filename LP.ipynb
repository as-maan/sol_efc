```python
import read

shot = 192974

nodes, labels, ntimes, pnums, R, Z, TEMP, DENSITY, JSAT, POT = read.shelf_probes(shot)
# display as a pandas dataframe

import pandas as pd
# list the length TEMP[0][node(idx)],

lengths = [],[],[],[]
for i,node in enumerate(nodes):
    lengths[0].append(len(TEMP[0][i]))
    lengths[1].append(len(DENSITY[0][i]))
    lengths[2].append(len(JSAT[i][0]))
    lengths[3].append(len(POT[0][i]))


df = pd.DataFrame({'R': R, 'Z': Z, 'NODES': nodes, 'LABELS': labels, 'NTIMES': ntimes, 'PNUMS': pnums,
                   'TEMP': lengths[0], 'DENSITY': lengths[1], 'JSAT': lengths[2], 'POT': lengths[3]})
df
```

```python
# make a heat map with R on y axis, time on x axis and pot as color

import matplotlib.pyplot as plt
import numpy as np
import plot_tools as plot
from scipy.signal import savgol_filter
from scipy.interpolate import interp1d

fig, ax = plt.subplots(1,1,figsize=(15,5))

v_float = []
t_max, t_min = [],[]
for i,r in enumerate(R): 
    v = POT[0][i].data
    time = POT[0][i].dim_0.data
    t_max.append(max(time))
    t_min.append(min(time))
span = [min(t_min),max(t_max)]
time = np.linspace(span[0],span[1],10000)
for i,r in enumerate(R): 
    v = POT[0][i].data
    t = POT[0][i].dim_0.data
    # interpolate v(t) to v(time)
    #v_int = np.interp(time,t,v)
    v_int = interp1d(t,v,kind='linear',fill_value='extrapolate')(time)
    # smooth v(time)
    v_smooth = savgol_filter(v_int, 201, 3)
    v_float.append(v_smooth)

# interpolate in the r direction as well 
r_int = np.linspace(min(R),max(R),100)
v_float = np.array(v_float)
v_float = v_float.transpose()
v_float = v_float.tolist()

for i,v in enumerate(v_float):
    #v_float[i] = np.interp(r_int,R,v,)
    v_f = interp1d(R,v,kind='linear',fill_value='extrapolate')(r_int)
    v_float[i] = savgol_filter(v_f, 75, 3)
v_float = np.array(v_float)
v_float = v_float.transpose()


ax.pcolormesh(time,r_int,v_float,cmap='seismic',vmin=-40,vmax=40)
```

```python
# make a heat map with R on y axis, time on x axis and pot as color

import matplotlib.pyplot as plt
import numpy as np
import plot_tools as plot
from scipy.signal import savgol_filter
from scipy.interpolate import interp1d

fig, ax = plt.subplots(1,1,figsize=(15,5))

v_float = []
t_max, t_min = [],[]
v_raw = []
for i,r in enumerate(R): 
    v = JSAT[i][0].data
    time = JSAT[i][0].dim_0.data
    t_max.append(max(time))
    t_min.append(min(time))
    
span = [min(t_min),max(t_max)]
time = np.linspace(span[0],span[1],10000)
for i,r in enumerate(R): 
    v = JSAT[i][0].data
    t = JSAT[i][0].dim_0.data
    # interpolate v(t) to v(time)
    #v_int = np.interp(time,t,v)
    v_int = interp1d(t,v,kind='linear',fill_value='extrapolate')(time)
    v_raw.append(v_int)
    # smooth v(time)
    v_smooth = savgol_filter(v_int, 201, 3)
    v_float.append(v_smooth)

# interpolate in the r direction as well 
r_int = np.linspace(min(R),max(R),100)

v_float = np.array(v_float)
v_raw = np.array(v_raw)

v_float = v_float.transpose()
v_raw = v_raw.transpose()

v_float = v_float.tolist()
v_raw = v_raw.tolist()

for i,(v,raw) in enumerate(zip(v_float,v_raw)):
    #v_float[i] = np.interp(r_int,R,v,)
    v_f = interp1d(R,v,kind='linear',fill_value='extrapolate')(r_int)
    v_raw[i] = interp1d(R,raw,kind='linear',fill_value='extrapolate')(r_int)
    v_float[i] = savgol_filter(v_f, 75, 3)
v_float = np.array(v_float)
v_raw = np.array(v_raw)

v_float = v_float.transpose()
v_raw = v_raw.transpose()


ax.pcolormesh(time,r_int,v_float,cmap='plasma',vmin=0,vmax=12)
```
