```python
import omfit_classes.omfit_mds as omds

shot = 192974
tree = omds.OMFITmds('DIII-D','Langmuir',shot=shot)
prbs = []
for key in tree.keys():
    if key.split('_')[0] == 'PROBE':
        r = omds.OMFITmdsValue(server='atlas.gat.com:8000',treename='LANGMUIR',shot=shot,TDI=f'\\Langmuir::TOP.{key}.R')()
        z = omds.OMFITmdsValue(server='atlas.gat.com:8000',treename='LANGMUIR',shot=shot,TDI=f'\\Langmuir::TOP.{key}.Z')()
        if r is not None:
            if r[0] > 1:
                label = omds.OMFITmdsValue(server='atlas.gat.com:8000',treename='LANGMUIR',shot=shot,TDI=f'\\Langmuir::TOP.{key}.LABEL')()[0]
                ntimes = omds.OMFITmdsValue(server='atlas.gat.com:8000',treename='LANGMUIR',shot=shot,TDI=f'\\Langmuir::TOP.{key}.NTIMES')()[0]
                pnum = omds.OMFITmdsValue(server='atlas.gat.com:8000',treename='LANGMUIR',shot=shot,TDI=f'\\Langmuir::TOP.{key}.PNUM')()[0]
                prbs.append([key,label,ntimes,pnum,r[0],z[0]])
# arrange prb in prbs in ascending r 
prbs = sorted(prbs,key=lambda x: x[4])
# display as a pandas dataframe
import pandas as pd

df_omfit = pd.DataFrame(prbs,columns=['NODENAME','Probe label','ntimes','pnum','r [m]','z[m]'])
# add header
df_omfit = df_omfit.style.set_caption('From omfit_classes.omfit_mds.OMFITmdsValue')
```

```python
# read idl sav file
import scipy.io as sio
import pandas as pd

lp = sio.readsav('lp_192974.sav')['foo']
a = []
for tmp in lp[0]:
    a.append(tmp)
prbs = []
for prb in a[2]:
    prbs.append(prb)
label, pnum, r, z, ntimes = [],[],[],[],[]
for prb in prbs:
    # convert byte string to string
    l = prb[1].decode('utf-8')
    label.append(l)
    pnum.append(prb[3])
    r.append(prb[4])
    z.append(prb[5])
    ntimes.append(prb[2])

# zip and sort by r
prbs = zip(label,pnum,r,z,ntimes)
prbs = sorted(prbs,key=lambda x: x[2])
label,pnum,r,z,ntimes = zip(*prbs)
# display as a pandas dataframe
df_idl = pd.DataFrame({'label':label,'ntimes':ntimes,'pnum':pnum,'r':r,'z':z})
df_idl = df_idl.style.set_caption('idl sav file compiled using `/fusion/projects/diagnostics/langmuir/mdsplus/get_lp.pro`')
```

```python
# display the two pandas dataframes side by side

from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
display_side_by_side(df_omfit,df_idl)
```
